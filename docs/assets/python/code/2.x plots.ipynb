```python
# Define convenience functions for setting up and completing figures
def figure_setup():
    
    fig = plt.figure(figsize=(6,4)) # 6x4 is possibly too bit, maybe make it 4x3 instead...
    ax = fig.add_subplot()

    # The following is now set in the matplotlibrc file
    #for direction in ["right","top"]:
    #    ax.spines[direction].set_visible(False)
    
    for direction in ["left","bottom"]:
        ax.spines[direction].set_position('zero')

    return ax

def figure_complete(ax,filename=None,svg_only=False):
    
    plt.tight_layout() # must come before grid(True) (if set manually)
    #ax.grid(True) # now set in matplotlibrc file
    
    # changes to axis limits have been removed...
    
    # optionally alter the axis ticks (remove zeros)
    ax.set_xticks([x for x in ax.get_xticks() if x!=0]) 
    ax.set_yticks([y for y in ax.get_yticks() if y!=0]) 
    
    #Commented out. I think this makes the view similar across the 3 types - geogebra doesnt label
    # Add x,y labels...
    #ax.text(0.00,1.02, r'$y$',transform=BlendedGenericTransform(ax.transData,ax.transAxes),ha='right',va='bottom')
    #ax.text(1.02,0.00, r'$x$',transform=BlendedGenericTransform(ax.transAxes,ax.transData),ha='left',va='top')
    # Optionally, add an origin O
    #ax.text(0.0,0.0, r'$\,O$',ha='left',va='bottom')
    
    if filename is not None:
        if not filename.startswith('../assets/'):
            filename = '../assets/'+filename
        plt.savefig(filename+'.svg',format='svg',bbox_inches='tight',pad_inches=0.03)
        if not svg_only:
            plt.savefig(filename+'.pdf',format='pdf',bbox_inches='tight',pad_inches=0.03)
    
    plt.show()

def figure_setup_2x2():
    
    fig = plt.figure(figsize=(7.2,4.8)) # Allow this to be a little bigger?
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    ax3 = fig.add_subplot(223)
    ax4 = fig.add_subplot(224)
    axes = [ax1,ax2,ax3,ax4]

    for ax in axes:
        # The following is now set in the matplotlibrc file
        #for direction in ["right", "top"]:
        #    ax.axis[direction].set_visible(False)

        for direction in ["left","bottom"]:
            ax.spines[direction].set_position('zero')

    return axes

def figure_complete_2x2(axes,filename=None,svg_only=False):
    
    plt.tight_layout() # must come before grid(True) (if set manually)
    
    for ax in axes:
        # The following is now set in the matplotlibrc file
        #ax.grid(True) # must come after tight_layout()
    
        # changes to axis limits have been removed...
    
        # optionally alter the axis ticks (remove zeros)
        ax.set_xticks([x for x in ax.get_xticks() if x!=0]) 
        ax.set_yticks([y for y in ax.get_yticks() if y!=0]) 
    
        # Add x,y labels...
        ax.text(0.00,1.02, r'$y$',transform=BlendedGenericTransform(ax.transData,ax.transAxes),ha='right',va='bottom')
        ax.text(1.02,0.00, r'$x$',transform=BlendedGenericTransform(ax.transAxes,ax.transData),ha='left',va='top')
        # Optionally, add an origin O
        ax.text(0.0,0.0, r'$\,O$',ha='left',va='bottom')
        
    plt.subplots_adjust(wspace=0.10, hspace=0.20)

    if filename is not None:
        if not filename.startswith('../assets/'):
            filename = '../assets/'+filename
        plt.savefig(filename+'.svg',format='svg',bbox_inches='tight',pad_inches=0.03)
        if not svg_only:
            plt.savefig(filename+'.pdf',format='pdf',bbox_inches='tight',pad_inches=0.03)
    
    plt.show()
```

```python
# Figures for 2.0 Linear functions
#
x = np.linspace(-5.0,5.0,129)
ax = figure_setup()
ax.plot(x, x)
figure_complete(ax,'2.0 x')
#
x = np.linspace(-5.0,5.0,129)
ax = figure_setup()
ax.plot(x, 4*x+5)
figure_complete(ax,'2.0 linear example 1')
#
x = np.linspace(-5.0,5.0,129)
ax = figure_setup()
ax.plot(x, 4*x+5)
ax.set_aspect(1.0)
ax.set_ylim(-8.0,12.0)
figure_complete(ax,'2.0 linear example 1 alt')
#
x = np.linspace(-5.0,5.0,129)
ax = figure_setup()
ax.plot(x, -2*x-4)
figure_complete(ax,'2.0 linear example 2')
#
x = np.linspace(-5.0,5.0,129)
ax = figure_setup()
ax.plot(x, np.abs(x))
figure_complete(ax,'2.0 piecewise example 1')
#
```
